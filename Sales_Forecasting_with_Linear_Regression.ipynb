# Sales Forecasting with Linear Regression

This sales forecasting tutorial compares actual sales and forecasted sales using different metrics such as mean squared error, mean absolute error, and $R^2$ score. For traning we will use Linear Regression.



# Import Packages

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

# Data

```python
store_sales = pd.read_csv("train.csv")
store_sales.head(10)
```

Check for missing values

```python
store_sales.info() # No missing values
```

Dropping store and item columns

```python
store_sales = store_sales.drop(['store', 'item'], axis=1)
store_sales.info()
```

Converting date from object datatype to daytime datatype

```python
store_sales['date'] = pd.to_datetime(store_sales['date'])
store_sales.info()
```

**Goal**: Train a regression model to predict sales for the next month.

```python
# Date to month period and grouping by item
store_sales['date']= store_sales['date'].dt.to_period('M')
monthly_sales = store_sales.groupby('date').sum().reset_index()
```

```python
# Convert the resulting date column to timestamp datatype
monthly_sales['date'] = monthly_sales['date'].dt.to_timestamp()
monthly_sales.head(10)
```

Monthly Sales by Item

```python
plt.figure(figsize=(15,5))
plt.plot(monthly_sales['date'], monthly_sales['sales'])
plt.xlabel('Date')
plt.ylabel("Sales")
plt.title('Monthly Customer Sales')
plt.show()
```

Call the difference on the sales columns to make the sales data stationary

```python
monthly_sales['sales_diff'] = monthly_sales['sales'].diff()
monthly_sales = monthly_sales.dropna()
monthly_sales.head(10)
```

```python
plt.figure(figsize=(15,5))
plt.plot(monthly_sales['date'], monthly_sales['sales'])
plt.xlabel('Date')
plt.ylabel("Sales")
plt.title('Monthly Customer Sales Difference')
plt.show()
```

Dropping off sales and date

```python
supervised_data = monthly_sales.drop(['date', 'sales'], axis =1)
```

Preparing the supervised data

```python
for i in range(1,13):
    col_name = "month_" + str(i)
    supervised_data[col_name] = supervised_data['sales_diff'].shift(i)
supervised_data = supervised_data.dropna().reset_index(drop=True)
supervised_data.head(10)
```

# Train and Test Data


```python
train_data = supervised_data[:-12]
test_data = supervised_data[-12:]
print("Train Data Shape", train_data.shape)
print("Test Data Shape", test_data.shape)
```

```python
scaler = MinMaxScaler(feature_range=(-1,1))
scaler.fit(train_data)
train_data = scaler.transform(train_data)
test_data = scaler.transform(test_data)
```

```python
X_train, y_train = train_data[:,1:], train_data[:,0:1]
X_test, y_test = test_data[:,1:], test_data[:,0:1]
y_train = y_train.ravel()
y_test = y_test.ravel()
print("X_train Shape: ", X_train.shape)
print("y_train Shape: ", y_train.shape)
print("X_test Shape: ", X_test.shape)
print("y_test Shape: ", y_test.shape)
```

Make prediction dataframe to merge the predicted sales price of all trained algorithms

```python
sales_dates = monthly_sales['date'][-12:].reset_index(drop=True)
predict_df = pd.DataFrame(sales_dates)
```

```python
actual_sales = monthly_sales['sales'][-13:].to_list()
actual_sales
```

To create the linear regression model and predicted output.

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
lr_pre = lr_model.predict(X_test)
```

```python
lr_pre = lr_pre.reshape(-1,1)
# This is a set matrix - contains the input feature of the test data, and also the predicted output
lr_pre_test_set = np.concatenate([lr_pre, X_test], axis = 1)
lr_pre_test_set = scaler.inverse_transform(lr_pre_test_set)
```

```python
result_list = []
for index in range(0, len(lr_pre_test_set)):
  result_list.append(lr_pre_test_set[index][0] + actual_sales[index])
lr_pre_series = pd.Series(result_list, name = "Linear Prediction")
predict_df = predict_df.merge(lr_pre_series, left_index = True, right_index = True)
predict_df
```

```python
lr_mse = np.sqrt(mean_squared_error(predict_df['Linear Prediction'], monthly_sales['sales'][-12:]))
lr_mae = mean_absolute_error(predict_df['Linear Prediction'], monthly_sales['sales'][-12:])
lr_r2 = r2_score(predict_df['Linear Prediction'], monthly_sales['sales'][-12:])
print("Linear Regression MSE:", lr_mse)
print("Linear Regression MAE:", lr_mae)
print("Linear Regression R^2:", lr_r2)
```

Visualization of the prediction vs actual sales


```python
plt.figure(figsize=(15,5))
# Actual sales
plt.plot(monthly_sales['date'], monthly_sales['sales'])
# Predicted sales
plt.plot(predict_df['date'], predict_df['Linear Prediction'])
plt.title("Customer sales Forecast using LRM")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend(["Actual Sales", "Predicted Sales"])
```
